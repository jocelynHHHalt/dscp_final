## Objective 1: Analyze the total purchase amounts across different seasons

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = 'shopping_trends.csv'  
data = pd.read_csv(file_path)
```

```python
# Group data by Season and Category 
detailed_summary = data.groupby(['Season', 'Category']).agg(
    Total_Purchase_Amount=('Purchase Amount (USD)', 'sum')
).reset_index()

# grouped bar chart
sns.catplot(
    data=detailed_summary,
    x='Season',
    y='Total_Purchase_Amount',
    hue='Category',
    kind='bar',
    height=6,
    aspect=1.5
)

# title and labels
plt.title('Total Purchase Amount by Season and Category')
plt.xlabel('Season')
plt.ylabel('Total Purchase Amount (USD)')

plt.show()
```

## Objective 2: Average Purchase Amount Trends 

```python
# Group data by Season and Category to calculate the average purchase amount
average_summary = data.groupby(['Season', 'Category']).agg(
    Average_Purchase_Amount=('Purchase Amount (USD)', 'mean')
).reset_index()

# Create the grouped bar chart for Average Purchase Amount Trends
sns.catplot(
    data=average_summary,
    x='Season',
    y='Average_Purchase_Amount',
    hue='Category',
    kind='bar',
    height=6,
    aspect=1.5
)

# Add title and labels
plt.title('Average Purchase Amount by Season and Category')
plt.xlabel('Season')
plt.ylabel('Average Purchase Amount (USD)')

# Show the plot
plt.show()
```
